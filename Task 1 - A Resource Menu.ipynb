```python
#obtaining information about computational resources
#construct a menu to give options
#Get the percentage usage of each CPU core over 10seconds and prints this out
#Get the amount of RAM currently being used and print this out in gigabytes
#Get the name and version of the operating system and print this out
#Check whether the computer is currently busy defined as both CPU and RAM usage being over 50%
#print whether computer is busy or not bus
#Exit


#import psutil library to retrieve information on runnung processes and system utilization
import psutil
import platform #to retrieve info of the system os and release


    
#initialize option
option = 0

#using the while loop, we can have it run to and meet all if functions
while option != 5:
         #using the print function, all the options are listed for user to selct fro
    option = int(input ("""       *******Resource Control********* 
                            [1] The percentage usage of each CPU core over 10secs
                            [2] The amount of RAM currently being used
                            [3] The name and version of the operating system
                            [4] Is the computer currently busy:
                            [5] Exit
                            
                            Type in a select number to choose a desired function
    
                        """))
    
        
            

 #statement to get CPU percentage over 10secs 
    if option == 1:
        usage = psutil.cpu_percent(10, True) #10 here represents usage over 10secs and True checks for each CPU core
        print("The percentage of cores are " + str(usage) + "%")
       
                
    
#statement to get RAM currently used
    elif option == 2:
        mem_usage = psutil.virtual_memory()
        ram_usage = mem_usage.available
        for i in  range(30): #conversion of RAM usage from bytes to GB, diving by 2 thirty times
             ram_usage = ram_usage /2
        print("The amount of RAM beng currently used is  " + str(ram_usage) + " GB of RAM")
        
        

    elif option == 3:
        print("The name of the operating system is " + platform.system() + ", version: " + platform.release())
      
    
    elif option == 4:
        cpu_usage = psutil.cpu_percent(1, False) #1 represents usage ver 1sec and Flase checks for all CPU cores
        ram_usage = psutil.virtual_memory()[2] #[2]is to call up just the % of ram usage from various statistics of the RAM
        if cpu_usage > 50 and ram_usage > 50:
            print("CPU usage is " + str(cpu_usage) + "% and RAM usage is " + str(ram_usage) + "%, therefore computer is busy")
        else:
            print("CPU usage is " + str(cpu_usage) + "% and RAM usage is " + str(ram_usage) + "%, therefore computer is not busy")
    
    elif option == 5:
        exit
        print("Exit")
     
    
    else:
        print("Please enter valid selection ") 
        
    
```
