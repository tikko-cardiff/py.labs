```python
#code must run forever continously printing information every 20secs
#the percentage CPU usage
#the number of processes running on the computer
#differences in the number of processes must be printed each time
#if the CPU usage has increased by 10% from 20scs earlier, print name of any process 
#using over 1% of total CPU. If none, print a message to state this


import psutil #to import the running of computer processes
import time #to import time lapse


while True:
    pids1 = len(psutil.pids()) #set to check process id 
    usage1 = psutil.cpu_percent(1, False) #checking for cpu usage over 1sec
    print("Your CPU usage is currently " + str(usage1) + "% and there are " + str(pids1) + " process running on this machine")

    
    print("Please wait 20 secs")
    time.sleep(5)#to set a time lapse of 20seconds
    pids2 = len(psutil.pids())
    usage2 = psutil.cpu_percent(1, False)
    print("2. Your CPU usage is currently " + str(usage2) + "% and there are " + str(pids2) + " process running on this machine")

    
    if pids1 < pids2: 
        pids3 = pids2 - pids1
        print("Processes has increased from " + str(pids1) + " processes to " + str(pids2) + " processes with " + str(pids3) + " more processe(s) runnung than before")
    
    elif pids1 > pids2:
        pids3 = pids1 - pids2
        print("Proccesss has decreased from " + str(pids1) + " processes to " + str(pids2) + " processes with " + str(pids3) + " less processe(s) running")
    
    else:
        print("Processes have stayed same at " + str(pids1) + " processes") 
    
    
    if usage1 < usage2: #to check if usage has increased
        usage3 = float(((usage2 - usage1)/ usage2)* 100) #to calculate percentage increase in usage
        if usage3 > 10  or usage3 == 10:
            p_name = []
            for process in psutil.process_iter(): #to check CPU time use by the process
                p = process.as_dict()
                if p["cpu_percent"] > 1:
                    p_name.append(p["name"])
            if len(p_name) > 0:
                print(p_name)
            else:
                print("None of the processes are using over 1%")
```
